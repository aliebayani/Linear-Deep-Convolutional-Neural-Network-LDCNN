```python
import numpy as np
import matplotlib.pyplot as plt

# Analog ECG signal (example data)
analog_ecg = [0.5, 0.7, 0.9, 0.6, 0.3, 0.2, 0.4, 0.8, 1.0, 0.5]

# Sampling rate (samples per second)
sampling_rate = 100  # Example: 100 samples per second

# Digital ECG signal (empty list)
digital_ecg = []

# Digitization process
for i in range(len(analog_ecg)):
    # Sampling: Take a sample at regular intervals
    if i % int(1000/sampling_rate) == 0:  # Assuming 1000 Hz analog signal
        sample = analog_ecg[i]

        # Quantization: Round or truncate the sample to a discrete level
        quantized_sample = round(sample, 2)  # Example: Rounded to 2 decimal places

        # Add the quantized sample to the digital ECG signal
        digital_ecg.append(quantized_sample)

# Plotting the results
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.plot(analog_ecg)
plt.title('Analog ECG Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.stem(digital_ecg, use_line_collection=True)
plt.title('Digital ECG Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Analog ECG signal (example data)
time = np.linspace(0, 1, 1000)  # Time array for analog ECG signal (1 second duration)

# Create the ECG signal with QRS complexes
analog_ecg = np.sin(2 * np.pi * 5 * time)  # Example: A sinusoidal waveform
qrs_amplitude = 0.5  # Amplitude of QRS complexes
qrs_duration = 0.05  # Duration of QRS complexes (in seconds)
qrs_start_time = 0.2  # Start time of the first QRS complex (in seconds)
qrs_end_time = qrs_start_time + qrs_duration

# Add QRS complexes to the ECG signal
qrs_indices = np.where((time >= qrs_start_time) & (time < qrs_end_time))[0]
analog_ecg[qrs_indices] += qrs_amplitude

num_samples = len(analog_ecg)  # Number of samples in the analog ECG signal

# Sampling rate (samples per second)
sampling_rate = 200  # Example: 200 samples per second

# Digital ECG signal (empty list)
digital_ecg = []

# Digitization process
for i in range(num_samples):
    # Sampling: Take a sample at regular intervals
    if i % int(1000/sampling_rate) == 0:  # Adjusted for the decreased sampling rate
        sample = analog_ecg[i]

        # Quantization: Round or truncate the sample to a discrete level
        quantized_sample = round(sample, 2)  # Example: Rounded to 2 decimal places

        # Add the quantized sample to the digital ECG signal
        digital_ecg.append(quantized_sample)

# Plotting the results
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.plot(time, analog_ecg)
plt.title('Analog ECG Signal with QRS Complexes')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.stem(np.arange(len(digital_ecg)), digital_ecg, linefmt='k-', markerfmt='ro', basefmt='k-', label='Digital ECG Signal')
plt.title('Digital ECG Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
